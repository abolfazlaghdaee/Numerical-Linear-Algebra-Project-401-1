```python

def genTriplematrix(A):

    p = np.identity(10)
    # print(p)
    for i in range(1,6):
        c= A[i:, i-1, np.newaxis]
        # print(c)
        h = gh.genUh(c)
        # print(h)
        H= Hbyh.gen_H(h)
        # print(H)
        p =H @ p
        
        
        
        
        
        A = H @ A @ H.T
 
    return  A ,p
      

        
```
