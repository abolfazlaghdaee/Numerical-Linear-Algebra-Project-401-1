```python

def gen_U(a):
  

    norm2 = 0
    t = a.shape[0]
    
    l = [[0] for i in range(t)]
    l[0] = [1]
    for i in a:
        norm2 += i**2
        
    norm2 = math.sqrt(norm2)    
        
    if a[0] <0:
           u=   a - norm2 * np.array(l)
    elif a[0] >= 0:
        u = a+ norm2 * np.array(l)
        
    i = np.identity(t) 
    h = i -  (2*(u@ u.T)/ (u.T @ u))
    return h
```

```python
def gen_H(a):
    n = a.shape[0]
    H = a
    
    
    for i in range(n, 4):
        c = np.array([[0 for i in range(n)]])
        c2 = np.array([ 0 for i in range(n+1)])

        f = np.append(c , H, axis =0)
        H = np.insert(f, 0 , c2, axis = 1)
        H[0,0] =1 


        if H.shape[0] == 4:
            return H

        else: 
            return gen_H(H)

        
   
```
